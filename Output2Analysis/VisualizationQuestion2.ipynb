```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Question2Output.csv') # Reading Data
```

```python
data
```

```python
Confirmed = data[["date","confirmed_rate"]]
```

```python
Confirmed.head()
```

```python
Death = data[["date","death_rate"]]
```

```python
Death.head()
```

```python
fig = plt.figure(figsize = (40, 15))
 
# creating the bar plot
plt.bar(Confirmed["date"], Confirmed["confirmed_rate"], color ='black',
        width = 0.3)
 
plt.xlabel("Date",fontsize=30)
plt.ylabel("Percentage Increase",fontsize=30)
plt.title("US- Confirmed Covid Cases Analysis",fontsize=50)
plt.xticks(rotation=90)
plt.xticks(color='blue')
plt.yticks(color='blue')
plt.show()
plt.savefig('US-ConfirmedCasesBarGraph.png')
```

```python
fig = plt.figure(figsize = (40, 15))
 
# creating the bar plot
plt.bar(Death["date"], Death["death_rate"], color ='black',
        width = 0.3)
 
plt.xlabel("Date",fontsize=30)
plt.ylabel("Percentage Increase",fontsize=30)
plt.title("US- Death Cases Analysis",fontsize=50)
plt.xticks(rotation=90)
plt.xticks(color='blue')
plt.yticks(color='blue')
plt.show()
plt.savefig('US-DeathCasesBarGraph.png') 
```

```python
#Plotting Line Graph

Confirmed.plot(x="date", y=["confirmed_rate" ], kind="line", figsize=(10, 7))
plt.xlabel("Date",fontsize=20)
plt.ylabel("Percentage Increase",fontsize=20)
plt.title("US- Confirmed Covid Cases Analysis",fontsize=30)

plt.savefig("US-ConfirmedCasesLineGraph.png")
```

```python
 #Plotting Line Graph

Death.plot(x="date", y=["death_rate" ], kind="line", figsize=(10, 7))
plt.xlabel("Date",fontsize=20)
plt.ylabel("Percentage Increase",fontsize=20)
plt.title("US- Death Covid Cases Analysis",fontsize=30)

plt.savefig("US-DeathCasesLineGraph.png")
```
