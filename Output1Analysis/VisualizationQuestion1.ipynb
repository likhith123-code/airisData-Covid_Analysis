```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
germany = pd.read_csv("germany_data.csv")
japan = pd.read_csv("japan_data.csv")
```

```python
germany.head()
```

```python
japan.head()
```

```python
# Confirmed Rate Analyis Between Germany and Japan
#Plotting Line Graph

germany.plot(x="date", y=["confirmed_rate","death_rate","recovered_rate"], kind="line", figsize=(10, 7))
plt.xlabel("Date",fontsize=18)
plt.ylabel("New cases",fontsize=18)
plt.title("Germany Cases Analysis - Lock Down on March 28",fontsize=22)
plt.show()
plt.savefig("GermanyCases.png")
```

```python
# Confirmed Rate Analyis Between Germany and Japan
#Plotting Line Graph

japan.plot(x="date", y=["confirmed-rate","death_rate","recovered_rate"], kind="line", figsize=(10, 7))
plt.xlabel("Date",fontsize=18)
plt.ylabel("New cases",fontsize=18)
plt.title("Japan Cases Analysis - No LockDown",fontsize=22)
plt.show()
plt.savefig("JapanCases.png")
```
