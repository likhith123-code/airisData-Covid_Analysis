```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("India_data.csv")
```

```python
data.head()
```

```python
# Confimed new cases if NoLD, only PD, PD + TotalLD

data.plot(x="date", y=["confimedDailyCases","ConfirmedIfPartialLD","ConfirmedIfNoLD"], kind="line", figsize=(10, 7))
plt.xlabel("Date",fontsize=18)
plt.ylabel("Confirmed New cases",fontsize=18)
plt.title("India Confirmed Cases Analysis",fontsize=22)
plt.show()
plt.savefig("ConfirmedCasesIndia.png")
```

```python
# Death new cases if NoLD, only PD, PD + TotalLD

data.plot(x="date", y=["deathsDailyCases","DeathsIfPartialLD","DeathsIfNoLD"], kind="line", figsize=(10, 7))
plt.xlabel("Date",fontsize=18)
plt.ylabel("Death New cases",fontsize=18)
plt.title("India Death Cases Analysis",fontsize=22)
plt.show()
plt.savefig("DeathCasesIndia.png")
```
